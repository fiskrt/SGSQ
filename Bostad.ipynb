```python
import json

json_data = json.loads(x.text)

res = []
for r in json_data['Result']:
    res.append(r)

res.sort(key=lambda x: x['CountInterest'])
```

```python
for r in res:
    l = geolocator.geocode(r['StreetName'] + ' ' + str(r['StreetNo']) + ' göteborg')
    if l:
        r['coords'] = l.latitude, l.longitude
        if l.latitude > 58 or l.latitude < 57 or l.longitude < 11 or l.longitude > 12.5:
            print(f"Not found in GBG: {r['StreetName']}")
    else:
        r['coords'] = None
```
